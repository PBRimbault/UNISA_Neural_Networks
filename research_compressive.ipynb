```python
# Research compressive strength testing
# Import the .csv and convert to NumPy array

import pandas as pd
import numpy as np


# Make NumPy values easier to read
np.set_printoptions(precision=3, suppress=True)

import tensorflow as tf
from tensorflow.keras import layers

names = ["Cement", "Aggregate", "FA", "Cyclodextrin", "Water", "Days", "Load 1", "Comp 1", "Load 2", "Comp 2", "Load 3", "Comp 3"]

compressive_dataset = pd.read_csv(
    '/content/drive/MyDrive/1 Projects/3_4 Research Project/Research data/compressive_all_data.csv', 
    names = ["Cement", "Sand", "Aggregate", "FA", "Cyclodextrin", "Water", "Days", "Comp 1", "Comp 2", "Comp 3"])

data_arr = compressive_dataset.to_numpy()

np.random.shuffle(data_arr)

X_values = data_arr[:,:7]
# print(X_values)
y_values = data_arr[:,8:9]
# print(y_values)

# 1. Create the model
model_1 = tf.keras.Sequential([
    tf.keras.layers.Dense(1)
])

# 2. Compile the model
model_1.compile(loss=tf.keras.losses.mae,
                optimizer=tf.keras.optimizers.SGD(),
                metrics=["mae"])

# 3. Fit the model
model_1.fit(tf.expand_dims(X_values, axis=-1), y_values, epochs=100, verbose=0)
```

```python
model_1.summary()
```

```python
y_preds_1 = model_1.predict(X_values)

mae_1 = tf.metrics.mean_absolute_error(y_true=y_values,
                                     y_pred=tf.squeeze(y_preds_1))

mse_1 = tf.metrics.mean_squared_error(y_true=y_values,
                                     y_pred=tf.squeeze(y_preds_1))

mae_1, mse_1
```
